## Data Visualization using Seaborn Library

```python
#importing necessary packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.set_style('dark')
```

```python
#Scatterplot with additional slices by tweaking size,shape and color
sns.relplot(x='total_bill',y='tip',data=tips,col='time',hue='smoker',style='sex',size='size')
plt.show()
```

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
#line plot
sns.relplot(x='time',y='firing_rate',data=dots,hue='choice',size='coherence',col='align',kind='line',facet_kws=dict(sharex=False),style='choice')
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
#line plot with confidence interval (default)
sns.set_style('darkgrid')
sns.relplot(x='timepoint',y='signal', data=fmri,style='event',col='region',kind='line')
plt.show()
```

```python
#plotting regression

sns.set_palette('Accent')
sns.set_style('whitegrid')
#from tip dataset
sns.lmplot(x='tip',y='total_bill', data=tips,hue='sex',col='time',col_wrap=2)
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#more regression
sns.set_style('whitegrid')
sns.set_palette('rainbow')
sns.lmplot(x='sepal_length',y='sepal_width',data=iris,col='species',col_wrap=2)
plt.plot()
```

```python
#histogram
sns.histplot(x='petal_width',data=iris,bins=10,palette='rbg',kde=True)
plt.show()
```

```python
#multiple histogram 
sns.displot(x='total_bill',data=tips,col='time',kde=True)
plt.show()
```

```python
#probability distribution curve
sns.displot(x='total_bill',kind='ecdf',data=tips,col='sex',color='red')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#swarnplots for categorical distribution
sns.catplot(x='survived',y='fare',data=titanic,hue='sex',col='pclass',palette='cool_r')
plt.show()
```

```python
#violinplot
sns.violinplot(x='survived',y='fare',data=titanic)
plt.show()
```

```python
sns.catplot(x='survived',y='fare',data=titanic,kind='violin',hue='sex',col='embark_town',col_wrap=2)
```

```python
#boxplot
sns.catplot(x='survived',y='age',kind='box',data=titanic,palette='Accent',col='sex')
plt.show()
```

```python
#pairplot to show distrubution and relation of each numeric variable in a dataset
sns.pairplot(data=iris,hue='species',palette='husl')
plt.show()
```

```python
#focus on two varables
sns.jointplot(x='age',y='fare',data=titanic,hue='class')
plt.show()
```

```python
#heatmap of correlation matrix
sns.heatmap(iris.corr(),cmap='copper')
```
